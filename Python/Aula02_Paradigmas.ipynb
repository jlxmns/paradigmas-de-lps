```python
# Minhas Funções
# Júlio César Toscano Ximenes Jr.

def criaVetor(tamanho):
  vetor = [0] * tamanho
  return vetor

def populaVetor(vetor):
  for i in range(len(vetor)):
    vetor[i] = int(input("Digite um valor para a posição {} do vetor: ".format(i)))

def escolherPosicoes(opcoes):
  vetorOpcoes = []
  for i in range(opcoes):
    vetorOpcoes.append(int(input("Digite a posição {}: ".format(i))))
  return vetorOpcoes

def somaValores(vetor, opcoes):
  soma = 0
  for i in opcoes:
    soma += vetor[i]
  return soma

def exibirSoma(soma, opcoes):
  print("A soma dos valores nas posições ({}) é: {}".format(", ".join(map(str, opcoes)), soma))

def descobrirMaior(vetor):
  maior = vetor[0]
  for i in range(1, len(vetor)):
    if vetor[i] > maior:
      maior = vetor[i]
  return maior

def descobrirMenor(vetor):
  menor = vetor[0]
  for i in range(1, len(vetor)):
    if vetor[i] < menor:
      menor = vetor[i]
  return menor

def somarTodos(vetor):
  soma = 0
  for i in range(len(vetor)):
    soma += vetor[i]
  return soma

def gerarCartela():
  numeros = set()
  cartela = []

  for i in range(5):
    cartela.append([])
    for j in range(5):
      numero = random.randint(0, 99)
      while numero in numeros:
        numero = random.randint(0, 99)
      numeros.add(numero)
      cartela[i].append(numero)

  return cartela
```

```python
# lê um vetor de 8 posições e, em seuida, lê dois valores de X e Y quaisquer e correspondentes a duas posições no vetor.
# O programa então imprime a soma dos valores encontrados nas respectivas posições X e Y.
# Júlio César Toscano Ximenes Jr.

def main():
  vetor = criaVetor(8)

  populaVetor(vetor)

  opcoes = escolherPosicoes(3)

  soma = somaValores(vetor, opcoes)

  exibirSoma(soma, opcoes)

main()
```

```python
# Faça um programa em Python que receba do usuário um vetor com 10 posições.
# Em seguida, deverá ser impresso o maior e o menor elemento do vetor, além da
# soma de todos os elementos do vetor.
# Júlio César Toscano Ximenes Jr.

def main():
  vetor = [0] * 10

  populaVetor(vetor)
  maior = descobrirMaior(vetor)
  menor = descobrirMenor(vetor)
  soma = somarTodos(vetor)

  print("O maior valor do vetor é: {}".format(maior))
  print("O menor valor do vetor é: {}".format(menor))
  print("A soma de todos os valores do vetor é: {}".format(soma))

main()
```

```python
# Faça um programa em Python para gerar automaticamente números entre 0 e 99 de uma cartela de bingo.
# Sabendo que cada cartela deverá conter 5 linhas de 5 números.
# Gere esses dados de modo a não ter números repetidos dentro das cartelas.
# O programa deve exibir na tela a cartela gerada.
# Júlio César Toscano Ximenes Jr.

import random

def main():
  cartela = []
  for i in range(5):
    cartela.append(gerarCartela())

  print("Cartela de bingo:")
  for linha in cartela:
    print("\nCartela")
    for coluna in linha:
      print(coluna)

main()
```

```python
# Coleta dados do usuário: recebe uma lista de nomes, idades e telefones.
# Armazena dados em um dicionário: usa um dicionário para associar cada nome à sua idade
# Salva dados em um arquivo.txt: grava os dados armazenados no dicionário em um arquivo de texto
# Júlio César Toscano Ximenes Jr.

def coletar_dados():
  dados = []
  while True:
    nome = input("Digite o nome (ou 'sair' para encerrar): ")
    if nome.lower() == 'sair':
      break

    idade = input("Digite a idade: ")
    try:
      idade = int(idade)
    except ValueError:
      print("Idade deve ser um número inteiro. Tente novamente.")
      continue

    telefone = input("Digite o telefone: ")
    dados.append({'nome': nome, 'idade': idade, 'telefone': telefone})
  return dados

def salvar_arquivos(dados, nome_arquivo):
  with open(nome_arquivo, 'w') as arquivo:
    for item in dados:
      linha = f"Nome: {item['nome']}, Idade: {item['idade']}, Telefone: {item['telefone']}\n"
      arquivo.write(linha)

def main():
  dados = coletar_dados()
  salvar_arquivos(dados, 'dados.txt')
  salvar_arquivos(dados, 'dados.csv')
  salvar_arquivos(dados, 'dados.json')

main()
```
