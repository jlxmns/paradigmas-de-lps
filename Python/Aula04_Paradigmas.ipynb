```python
class Animal:
    def __init__(self, nome):
        self.nome = nome

    def emitir_som(self):
        pass

class Mamifero(Animal):
    def __init__(self, nome):
        super().__init__(nome)

    def amamentar(self):
        print(f"{self.nome} está amamentando.")

class Ave(Animal):
    def __init__(self, nome):
        super().__init__(nome)

    def voar(self):
        print(f"{self.nome} está voando.")

class Morcego(Mamifero, Ave):
    def __init__(self, nome):
        Mamifero.__init__(self, nome)
        Ave.__init__(self, nome)

    def emitir_som(self):
        return "Scree"

def main():
    morcego = Morcego("Morcego")
    morcego.amamentar()
    morcego.voar()
    print(morcego.emitir_som())

if __name__ == "__main__":
    main()
```

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        self.endereco = None

    def adicionar_endereco(self, endereco):
        self.endereco = endereco

    def mostrar_informacoes(self):
        print(f"Nome: {self.nome}, Idade: {self.idade}")
        if self.endereco:
            self.endereco.mostrar_endereco()
        else:
            print("Endereço não cadastrado.")

class Endereco:
    def __init__(self, rua, cidade, estado, cep):
        self.rua = rua
        self.cidade = cidade
        self.estado = estado
        self.cep = cep

    def mostrar_endereco(self):
        print(f"Rua: {self.rua}, Cidade: {self.cidade}, Estado: {self.estado}, CEP: {self.cep}")

class Empresa:
    def __init__(self, nome, cnpj):
        self.nome = nome
        self.cnpj = cnpj
        self.funcionarios = []

    def contratar_funcionario(self, funcionario):
        self.funcionarios.append(funcionario)

    def listar_funcionarios(self):
        print(f"Funcionários da empresa {self.nome}:")
        for funcionario in self.funcionarios:
            print(funcionario.nome)

def main():
    endereco1 = Endereco("Avavav avav", "xxXX", "EeEeE", "12345-678")
    pessoa1 = Pessoa("João", 30)
    pessoa1.adicionar_endereco(endereco1)

    endereco2 = Endereco("Rua qwfoiqbwofi", "Iiodqowi", "qwpibwqi", "12903-123")
    pessoa2 = Pessoa("Maria", 25)
    pessoa2.adicionar_endereco(endereco2)

    empresa = Empresa("Empresa ABC", "12.345.678/0001-90")

    empresa.contratar_funcionario(pessoa1)
    empresa.contratar_funcionario(pessoa2)

    empresa.listar_funcionarios()

if __name__ == "__main__":
    main()
```

```python
class Motor:
    def __init__(self, tipo, potencia):
        self.tipo = tipo
        self.potencia = potencia

    def ligar(self):
        print(f"O motor está ligado.")

    def desligar(self):
        print(f"O motor está desligado.")

class Pneu:
    def __init__(self, marca, tamanho):
        self.marca = marca
        self.tamanho = tamanho

    def inflar(self):
        print(f"O pneu está inflado.")

    def desinflar(self):
        print(f"O pneu está deflado.")

class Carro:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self.motor = Motor("Gasolina", 150)
        self.pneus = [Pneu("Pirelli", 18) for _ in range(4)]

    def ligar(self):
        self.motor.ligar()
        print(f"O seu {self.marca} {self.modelo} está pronto para rodar.")

    def desligar(self):
        self.motor.desligar()
        print(f"O {self.marca} {self.modelo} está desligado.")

    def trocar_pneu(self, indice, novo_pneu):
      self.pneus[indice] = novo_pneu
      print(f"Pneu {indice + 1} trocado com sucesso.")

def main():
    carro = Carro("Toyota", "Corolla")
    carro.ligar()
    carro.trocar_pneu(0, Pneu("Pirelli", 18))
    carro.desligar()

if __name__ == "__main__":
    main()
```
